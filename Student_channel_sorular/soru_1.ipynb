# Soru 1 --list comprehension ile
  Kullanıcıdan bir kelime alın ve bu kelimenin ilk harfinden başlayarak her seferinde bir harfi büyük harf olacak sekilde son harfe kadar işlemi tekrarlayan ve sonuçları liste içinde çıktı veren bir algoritma oluşturun.
Example:
['Ankara', 'aNkara', 'anKara', 'ankAra', 'ankaRa', 'ankarA']

```python
"""Kullanicidan bir kelime alin ve 
---bu kelimenin ilk harfinden başlayarak her seferinde bir harfi büyük harf olacak sekilde son harfe kadar 
işlemi tekrarlayan ve 
---sonuçlari liste içinde çikti veren bir algoritma oluşturun.
Example:
['Ankara', 'aNkara', 'anKara', 'ankAra', 'ankaRa', 'ankarA']"""


k_kelime = input("Lutfen bir kelime giriniz").lower()
kelime_list=[]
for i in range(len(k_kelime)) :
    x = k_kelime[:i] + k_kelime[i:i+1].upper() + k_kelime[i+1:]
    kelime_list.append(x)

print(kelime_list)        
```
